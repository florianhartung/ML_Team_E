# Labor Maschinelles Lernen

Das *Labor Maschinelles Lernen* 
* vertieft durch eine Projektarbeit die in der Vorlesung dargestellten Methoden und Konzepte. 
* erfordert die eigenständige, wissenschaftliche Bearbeitung einer realistische Problemstellung des maschinellen Lernens in Teamarbeit
* setzt auf die Verwendung von Jupyter Notebooks
* setzt auf die Verwendung von __PyTorch__ für tiefe neuronale Netzwerke
    * bei Bedarf können auch andere Framework wie TensorFlow verwendet werden




## Organisatorisches

* Die Bearbeitung des Labors wird in Gruppen aus drei Studierenden durchgeführt.
* Jede Gruppe erhält einen Buchstaben als Namen (`A`, `B`, `C`,...). 
* Ihre Aufgabenstellung erhalten Sie über das Moodle-Portal.  
* Die Aufgaben werden auf einer relativ großen Datenmenge ausgeführt. Dafür können sie eigenen Infrastrukturen verwenden oder Ressourcen der Hochschule zugreifen. Kontaktieren Sie dafür bitte den Laboringenieur Wassiliy Kaiser.  
* Für den sehr seltenen Fall, dass die Zusammenarbeit in einem Team nicht funktioniert und einer ihrer Teampartner sich nicht an einer Lösung beteiligt, steht es Ihnen frei, mit dem Dozenten Kontakt aufzunehmen, um eine Lösung zu suchen.


## Zu erbringende Leistungen

Ihre Prüfungsleistung beruht auf zwei verschiedenen Teilen:

1. Die Erstellung eines ausführbaren **Jupyter Notebooks**, das die Dokumentation zu Ihrer Lösung enthält. Hier dokumentieren Sie Ihr Projekt und erläutern die Schritte, die Sie während der Bearbeitung durchgeführt haben. 
        
2. Die **Präsentation der Ergebnisse** in der letzten Veranstaltung im Rahmen eines Kurzvortrags. 

Vergessen Sie nicht auch ihre Python Laufzeitumgebung, auf der Ihr Jupyter Notebook ausgeführt werden soll, zu beschreiben.    

### Jupyter Notebook - Abgabe

* Pro Gruppe wird *eine* gemeinsame Lösung eingereicht. 
* Pro Gruppe wird *eine* gemeinsame Bewertung durchgeführt.    
* Sie geben **pro Gruppe eine zip-Datei** ab 
    * Auf der obersten Ebene ihres Projekts befindet sich Ihr Jupyter Notebook.
    * Vergessen nicht, dass Sie zu Beginn des Notebook die Namen aller Teilnehmer mit Matrikelnummer angeben. 
    * WICHTIG: *das Notebook muss vor den Hochladen ausgeführt* werden und alle Ergebnisse und Plots sind enthalten.
    * Im Verzeichnis `src` müssen eigene Implementierungen der von Ihnen erstellten Python-Klassen und -Funktionen enthalten sein. Diese sollen dann durch das Notebook aufgerufen werden.  
    * Weitere notwendige Verzeichnisse können sie hinzufügen, wenn diese im Jupyter Notebook dokumentiert sind. 

* Die Abgabefrist für Ihre Aufgabenstellung wird im Moodle-Kurs angegeben. Laden Sie dort rechtzeitig Ihre zip-Datei hoch.
* Bennen Sie Ihr Notebook nach dem folgenden Schema: `ML-Labor-{Kursnummer}-Team-{Gruppenname}.zip` 
    * Beispiel: `ML-Labor-TINF19IT5-Team-E.zip` für ein Team mit dem Namen `E` aus dem Kurs `TINF19IT5`

## Bewertung

Sie haben zwei Leistungen zu erbringen. 
1. ein Jupyter-Notebook, das mit maximal 75 Punkten bewertet wird. 
2. eine Präsentation der Ergebnisse, die mit maximal 25 Punkten bewertet wird. Die Präsentation muss ebenfalls abgegeben werden. 

Beide Leistung ergeben die gesamten Prüfungsleistung für das *Labor Maschinelles Lernen*

Zusammen mit dem Ergebnis der Klausur ergibt sich die Gesamtnote für das Modul "Maschinelles Lernen". 
Dabei trägt die Klausur 75% und das Labor 25% zur Note bei. 

###  Jupyter Notebook - Bewertung und Anforderungen

* Dokumentieren Sie wissenschaftlich nachvollziehbar Ihre Codes, Visualisierungen, Referenzen, und Gedanken im Jupyter Notebook
    * Es ist nicht ausreichend irgendeine Lösung zu präsentieren. Sie müssen auch darstellen können wieso Sie eine Lösung gewählt haben.
    * Sie dürfen gerne auch in wissenschaftlichen Bibliotheken nach Möglichkeiten suchen. Vergessen Sie dann die Referenzen nicht.  
* Verwenden Sie zur Bearbeitung die in der Vorlesung dargestellten Methoden, Konzepte und Lösungsstrategien. 
    * Ordnen Sie die Aufgabe einer entsprechenden Kategorie zu (Bildverarbeitung, Sequenzanalyse, ...)
    * Wählen Sie als "Baseline" für die Performance evtl. einen der einfachen ML-Algorithmen aus. 
* Legen Sie Wert auf die Verwendung von tiefen neuronalen Netzen!  Verwenden Sie hierzu `python` und `PyTorch`.
    * Die Implementierung sollte mit diesen Tools realisiert sein. 
        * Alternativen sind möglich (siehe oben)
* Je nach Aufgabenstellung benötigen Sie noch weitere python-Bibliotheken oder zusätzliche Literatur bzw. Konzepte. Vergessen Sie nicht, diese Referenzen in Ihrer Lösung anzugeben. Erstellen Sie ggf. eine `requirements`-Datei, die zusätzliche Bibliotheken auflistet.
    * Der gesamte verwendete Code muss mit abgegeben werden. 
    * Scheuen Sie sich nicht, sich in neue Frameworks und Bibliotheken einzuarbeiten.
    * Das einfache Präsentieren einer gefundenen Lösung durch Copy&Paste reicht alleine nicht aus. Sie müssen den Ansatz der gefundenen Lösung verstanden haben und präsentieren können.  
* Bitte beachten Sie, dass es - wie in realen Anwendungsfällen üblich - nicht die eine perfekte Lösung gibt. Vielmehr sollen Sie ihr erworbenes Fachwissen, Ihre Problemlösungskompetenz und die Nachvollziehbarkeit Ihrer Überlegungen unter Beweis gestellt werden.

* In Ergänzung zu der eigentlichen Problemstellung (z.B. "Unterscheiden Sie Hunde von Katzen") sollten Sie u.A. folgende weitere Aspekte berücksichtigen und darstellen.
    * Analyse der Trainingsdaten
    * Auswahl, Bearbeitung und ggf. Transformation der Daten
    * Vergleich verschiedener Lösungsansätze und Modelle
    * Gegenüberstellung verschiedener Lösungsansätze
    * Überprüfen Sie die Implementierung und versuchen Sie auch Hyperparameter zu optimieren 
    * "Lessons Learned" - Darstellung von Herausforderungen und Ihren Lösungen
    * Ergänzen Sie nach Möglichkeit auch eigene Ideen und Fragestellungen

* Die Verwendung von geeigneten Visualisierungen mit denen Sie den Fortschritt ihrer Lernmodelle nachweisen wird empfohlen.
    * Überlegen Sie, welche Aussage Sie mit welcher Darstellungsweise zeigen möchten.

* Die Probleme sind komplex und hochdimensional und damit herausfordernd. 
    * Auch wenn Ihr Modell keine zufriedenstellende Ergebnisse (niedrige Accuray, keine Konvergenz) für das Training liefert, können Sie eine gute Bewertung erhalten, wenn Sie das Verhalten erklären oder begründen können. Es wird entsprechend Wert auf Ihr Verständnis gelegt. 

* Sie können maximal 75 Punkte erreichen.

### Präsentation - Bewertung und Anforderungen

In einer wissenschaftlichen Kurzpräsentation sollen Sie abschließend kurz Ihre Aufgabenstellung, Lösungsansätze und Ergebnisse vorstellen.

Gestalten Sie diese Präsentation so, dass ihre Kommilitonen aus Ihrer Präsentation das Wichtige und das Neue kennenlernen. Sollten Sie etwa
ein neues nicht in der Vorlesung vorgestelltes Verfahren verwenden, sollten sie die wesentlichen Konzepte vorstellen. 

- Fassen Sie die Ergebnisse und Überlegungen aus dem Jupyter Notebook kurz und prägnant zusammen.
- Die Zeitvorgabe von 20 Minuten ist dabei genau einzuhalten.
- Anschliessend stehen 5 Minuten für Diskussion zur Verfügung.  
- Teilen Sie sich die Präsentation zu ungefähr gleichen Redeanteilen auf.

* Sie können maximal 25 Punkte erreichen.
